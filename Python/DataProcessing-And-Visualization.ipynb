```python
# Q 3 answer
columns_to_work_upon = []

for column in df.columns:
    if df[column].isnull().any() or df[column].dtype != np.float64:
        columns_to_work_upon.append(column)

print("Columns that need to be worked upon:", columns_to_work_upon)
```

```python
# Q 4 answer
df.drop_duplicates(inplace=True)
print(df)
# there are in total 10841 rows but after dropping the duplicates there are only 10358 rows.
```

```python
# Q 5 answer
unique_categories = df['Category'].unique()
print(unique_categories)
```

```python
# Q 5 answer
# removing invalid categories
invalid_categories = ['1.9']
df = df[~df['Category'].isin(invalid_categories)]
print(df)
```

```python
# Q 6 answer 
# Check if there are missing values present in the column Rating, If any? drop them
missing_rating_value = df[df['Rating'].isnull()]

if missing_rating_value.shape[0] > 0:
    df = df.dropna(subset=['Rating'])
print(df)
```

```python
# Q 6 answer
# create a new column as 'Rating_category' by converting ratings 
# to high and low categories(>3.5 is high rest low)
df['Rating_category'] = 'low'
df.loc[df['Rating'] > 3.5, 'Rating_category'] = 'high'
print(df['Rating_category'])
```

```python
# Q 7 answer
rating_category_distribution = df['Rating_category'].value_counts()
print(rating_category_distribution)

# This tells us that there are 8012 high and 880 low categories in the "Rating_category" column.
# There are more high compared to low category data. This means that there are more ratings greater 
# than 3.5 compared to lower than 3.5 
```

```python
# Q 8 answer
# converting reviews column to numeric data
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
print(df['Reviews'])
```

```python
# Q 8 answer
# Converted the column 'Reviews' to numeric data.
Q1 = df['Reviews'].quantile(0.25)
Q3 = df['Reviews'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Reviews'] < lower_bound) | (df['Reviews'] > upper_bound)]

# Handle outliers using log transformation
df['Reviews'] = np.log1p(df['Reviews'])

print(df['Reviews'])
```

```python
# Q 9 answer
# this fuction returns 1 million if size value ends with 'M' and 
# returns 1 thousand if size value ends with 'K'. If it doesn't 
# have either value then returns None which should be 'Varies with device' value
def size_conversion(value):
    if value.endswith('M'):
        return float(value[:-1]) * 1e6
    elif value.endswith('K'):
        return float(value[:-1]) * 1e3
    else:
        return None

# call size_conversion function
df['Size'] = df['Size'].apply(size_conversion)

# dropping entries which are None which is equivalent to 'Varies with device' value
df = df[df['Size'].notnull()]
print(df['Size'])
```

```python
# Q 10 answer
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
print(df['Installs'])
```

```python
# Q 11 answer
df['Price'] = df['Price'].str.replace('[\$,]', '', regex=True)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
print(df['Price'])
```

```python
# Q 12 answer
drop_columns = ['App', 'Rating', 'Genres', 'Last Updated', 'Current Ver', 'Android Ver']
df.drop(drop_columns, axis=1, inplace=True)
```

```python
# Q 12 answer
print(df)
```

```python
# Q 13 answer
categorical_columns = ['Category']
# performing encoding for categorical_columns list given above and add it to the dataframe
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
print(df)
```

```python
# Q 14 Answer
# Segregating  the independent features
independent_features = df.drop(columns=['Rating_category'])
# Segregating the target feature (Rating_category)
target = df['Rating_category']


print(independent_features)
```

```python
# Q 15 Answer
from sklearn.model_selection import train_test_split
# Spliting the data into training and testing sets(80% training, 20% testing)
train_independent_data, test_independent_data, train_target, test_target = train_test_split(independent_features, target, test_size=0.2, random_state=42)
print(train_independent_data.shape)
print(test_independent_data.shape)
print(train_target.shape)
print(test_target.shape)
```

```python
# Q 16 Answer
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

train_independent_scaled = scaler.fit_transform(train_independent_data)
test_independent_scaled = scaler.transform(test_independent_data)
print(train_independent_scaled)
```
