```python
# Q 4 answer This is for finding the categorical count
categorical_summary = {}
for col in df.select_dtypes(include='object').columns:
    categorical_summary[col] = df[col].value_counts()

print("\nCount and Frequency of Categorical Columns:")
for col, summary in categorical_summary.items():
    print(f"\nColumn: {col}")
    print(summary)
```

```python
# Q 5 answer get the current datatype for published_at column
print("getting the current datatype for published_at column:")
print(df['published_at'].describe(datetime_is_numeric=True))

# converting the publised_at column to pandas datetime
print("\nconverting published_at column to pandas datetime")
df['published_at'] = pd.to_datetime(df['published_at'])
df['published_at'] = pd.to_datetime(df['published_at'], format =' %Y-%m-%d %H:%M:%S')
print(df.dtypes)
```

```python
# Q 6 answer. This question is asking to display the months only not month_name.
df['published_month'] = df['published_at'].dt.month
print(df['published_month'])
```

```python
# Q 7 answer
month_mapping_dict = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 
                 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
df['published_month'] = df['published_month'].replace(month_mapping_dict)
print(df['published_month'])
```

```python
# Q 8 as answer
total_video_count_by_published_month = df['published_month'].value_counts()
sorted_total_video_count_by_month = total_video_count_by_published_month.sort_values(ascending=False)
print(sorted_total_video_count_by_month)
```

```python
# Q 11 answer
maximum_likes = df.loc[df['likes'].idxmax(), 'title']
print("title of the video with maximum likes:", maximum_likes)


minimum_likes = df.loc[df['likes'].idxmin(), 'title']
print("title of the video with minimum likes:", minimum_likes)
```

```python
# Q 11 answer
"""
maximum_likes and minimum_likes variables hold the title of the video with maximum likes and minimum likes respectively.

This data provides insights into how the video is received by the audience, whether it is popular with the audience and 
what the content creator needs to post next time. 

If there number of likes are lower, it means the audience doesn't like the content of the video or the topic of the 
video since the video didn't connect with the audience. Maybe it is having controversial information 
for example: political or religious content. This data helps the content creator to improve or change their content next time.

If the likes are higher then the audience very much likes the video and this data can help the content creator to 
create more similar content.
"""
```

```python
# Q 12 answer
maximum_dislikes = df.loc[df['dislikes'].idxmax(), 'title']
print("title of the video with maximum dislikes:", maximum_dislikes)


minimum_dislikes = df.loc[df['dislikes'].idxmin(), 'title']
print("title of the video with minimum dislikes:", minimum_dislikes)
```

```python
# Q 12 answer
"""
maximum_dislikes and minimum_dislikes variables hold the title of the video with maximum dislikes and minimum dislikes respectively.

This data provides insights into how the video is received by the audience. 

If there number of dislikes are higher, it means the audience doesn't like the content of the video or the 
video is having controversial information for example: political or religious content. This data helps the 
content creator to improve their content next time.

If the dislikes are lower then the audience very much likes the video and this data can help the content creator to 
create more similar content.
"""
```

```python
# Q 13 answer
correlation_coefficient = df['view_count'].corr(df['dislikes'])
print("correlation coefficient of view_count and dislikes:", correlation_coefficient)

"""
Since the correlation_coefficient value is positive indicates a positive correlation between view_count and dislikes. 

This means the number of views did have effect on how many people disliked the video. 
More the number of audience who watched the video disliked the video because it might be 
controversial or didn't match their ideology. 

Even the plot chart given below matches with this conclusion where the the number of view_count 
increases the number of dislike increases respectively
"""
```

```python
# Q 13 answer
import matplotlib.pyplot as plt

plt.scatter(df['view_count'], df['dislikes'])
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.title('relationship between number of video views and dislikes')
plt.show()
```
